## **Data Standardization**

### Standardization kolom Sub Category

```python
# Menyamakan data menjadi huruf kapital
gc["Sub Category"] = gc["Sub Category"].str.strip()
gc["Sub Category"] = gc["Sub Category"].str.title()
gc['Sub Category'].head(5)
```

### Standardization kolom Price

```python
# Menghapus simbol dolar ($) pada kolom Price
gc["Price"] = gc["Price"].str.replace("$", "", regex=False)
gc['Price'].unique()[:10]
```

### Standardization Kolom Discount

```python
# Mengekstrak angka diskon pada kolom Discount
gc["Discount"] = gc["Discount"].replace({
    "No Discount": "0",
    "This item is not returnable.": "0"
})

gc["Discount"] = gc["Discount"].str.extract(r"(\d+\.?\d*)")
gc["Discount"].unique()[:10]
```

```python
gc[gc["Discount"].isna()]
```

### Standardization Kolom Rating

```python
# Mengubah nilai "No reviews" menjadi NaN
gc["Rating"] = gc["Rating"].replace("No Reviews", None)

# Memeriksa kembali nilai "No Reviews"
gc["Rating"].str.contains("No Reviews", na=False).sum()
```

### Standardization Kolom Title

```python
# Menghapus spasi ganda dan new line (\n)
gc["Title"] = gc["Title"].str.replace(r"\s+", " ", regex=True)
gc["Title"] = gc["Title"].str.replace("\n", " ", regex=False)
gc["Title"] = gc["Title"].str.strip()
gc["Title"].head()
```

### Standardization Kolom Currency

```python
# Mengubah simbol ($) menjadi USD
gc["Currency"] = gc["Currency"].replace("$", "USD")
```

### Standardization Kolom Feature

```python
# Menghapus new line (\n), spasi ganda, dan "
gc["Feature"] = gc["Feature"].str.replace("\n", " ", regex=False)
gc["Feature"] = gc["Feature"].str.replace('"', "", regex=False)
gc["Feature"] = gc["Feature"].str.replace(r"\s+", " ", regex=True)
gc["Feature"] = gc["Feature"].str.strip()
gc["Feature"].head()
```

### Standardization Kolom Product Description

```python
# Menghapus new line (\n) dan spasi ganda
gc["Product Description"] = gc["Product Description"].str.replace("\n", " ", regex=False)
gc["Product Description"] = gc["Product Description"].str.replace(r"\s+", " ", regex=True)
gc["Product Description"] = gc["Product Description"].str.strip()
gc["Product Description"].head()
```

## **Data Enrichment**

```python
# Membuat kolom Final_Price (harga setelah diskon)
gc["Final_Price"] = gc["Price"] - gc["Discount"]
gc["Final_Price"] = gc["Final_Price"].clip(lower=0)
```

```python
# Membuat kategori rating
gc["Rating_Category"] = pd.cut(
    gc["Rating_Value"],
    bins=[0, 2, 4, 5],
    labels=["Low", "Medium", "High"]
)
```

```python
# Membuat kolom Popularity
gc["Popularity"] = gc["Review_Count"]
```

```python
# Membuat kategori popularitas berdasarkan jumlah review
gc["Popularity_Category"] = pd.cut(
    gc["Review_Count"],
    bins=[0, 50, 200, 10000],
    labels=["Low", "Medium", "High"]
)
```

## **Data Aggregation**

#### Jumlah produk per kategori

```python
# Menghitung jumlah produk di setiap Sub Category
product_count = gc.groupby("Sub Category")["Title"].count()

print("Jumlah produk per kategori:")
print(product_count)
```

#### Rata-rata harga per kategori

```python
# Menghitung rata-rata harga produk di setiap kategori
avg_price = gc.groupby("Sub Category")["Price"].mean()

print("\nRata-rata harga per kategori:")
print(avg_price)
```

#### Rata-rata rating per kategori

```python
# Menghitung rata-rata rating produk di setiap kategori
avg_rating = gc.groupby("Sub Category")["Rating_Value"].mean()

print("\nRata-rata rating per kategori:")
print(avg_rating)
```

#### Total review per kategori

```python
# Mengetahui kategori mana yang paling populer
total_review = gc.groupby("Sub Category")["Review_Count"].sum()

print("\nTotal review per kategori:")
print(total_review)
```

#### Rata-rata diskon per kategori

```python
# Mengetahui kategori mana yang sering diskon
avg_discount = gc.groupby("Sub Category")["Discount"].mean()

print("\nRata-rata diskon per kategori:")
print(avg_discount)
```

#### Menggabungkan semua hasil agregasi ke satu tabel

```python
# Menggabungkan semua hasil agregasi ke satu tabel
summary = gc.groupby("Sub Category").agg({
    "Title": "count",
    "Price": "mean",
    "Rating_Value": "mean",
    "Review_Count": "sum",
    "Discount": "mean"
})

# Mengganti nama kolom agar lebih jelas
summary.columns = [
    "Total_Product",
    "Avg_Price",
    "Avg_Rating",
    "Total_Review",
    "Avg_Discount"
]

print("\nSUMMARY PER KATEGORI:")
print(summary)
```

# **Data Warehouse (Star Schema)**

### Membuat Dim_Category

```python
# Ambil kategori unik
dim_category = gc[["Sub Category"]].drop_duplicates().reset_index(drop=True)

# Tambahkan ID (primary key)
dim_category["Category_ID"] = dim_category.index + 1

# Urutkan kolom
dim_category = dim_category[["Category_ID", "Sub Category"]]

print(dim_category.head())
```

### Membuat Dim_Product

```python
# Ambil informasi produk
dim_product = gc[[
    "Title",
    "Feature",
    "Product Description",
    "Currency"
]].drop_duplicates().reset_index(drop=True)

# Tambahkan Product_ID
dim_product["Product_ID"] = dim_product.index + 1

# Urutkan kolom
dim_product = dim_product[[
    "Product_ID",
    "Title",
    "Feature",
    "Product Description",
    "Currency"
]]

print(dim_product.head())
```

```python
# Menghubungkan Category ke Product
gc = gc.merge(dim_category, on="Sub Category", how="left")

# Menghubungkan Product ke Data
gc = gc.merge(dim_product, on=["Title","Feature","Product Description","Currency"], how="left")
```

### Membuat Fact Table

```python
fact_table = gc[[
    "Product_ID",
    "Category_ID",
    "Price",
    "Discount",
    "Final_Price",
    "Rating_Value",
    "Review_Count"
]]

print(fact_table.head())
```
